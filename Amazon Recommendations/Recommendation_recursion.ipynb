```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as sp
import operator
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
%matplotlib inline
```

```python
def recommendations(df):
    df.columns = ['productid','product_title','price','userid','profileName','helpfulness','score','time','summary','review']
    df.drop(['price','profileName','time'],axis =1,inplace = True)
    
    print("------------------------------------------------------------------------------------------------------------")
    print("There are ",pd.DataFrame(df.product_title.value_counts()).shape[0]," unique products in the dataset")
    print("There are ",pd.DataFrame(df.userid.value_counts()).shape[0]," unique users in the dataset")
    
    df.groupby('productid').count()
    
    print("Number of Ratings for each product")
    num_rating = df.groupby('productid').count()['score']
    num_rating = pd.DataFrame(num_rating)
    num_rating.rename(columns = {'score':'num_ratings'},inplace = True)
    num_rating.sort_values(by = 'num_ratings',ascending = False)
    
    avg_rating = df.groupby('productid').mean()['score']
    avg_rating = pd.DataFrame(avg_rating)
    avg_rating.rename(columns = {'score':'avg_ratings'},inplace = True)
    avg_rating.sort_values(by = 'avg_ratings',ascending = False)
    
    popular_df = num_rating.merge(avg_rating ,on = 'productid')
    
    print(popular_df)
    
    print("-----------------------------------------------------------------------------------------------------------------------")
    # print("Providing a num_ratings limit")
    popular_df[popular_df['num_ratings']>2]
    x = df.groupby('userid').count()['score'] > 3  
    rated_users = x[x].index
    
    filtered_rating = df[df['userid'].isin(rated_users)]
    
    y = filtered_rating.groupby('productid').count()['score']>=0
    famous_products = y[y].index
    
    final_ratings = filtered_rating[filtered_rating['productid'].isin(famous_products)]
    
    pt = final_ratings.pivot_table(index='productid',columns='userid',values='score')
    pt.fillna(0,inplace = True)
    
    from sklearn.metrics.pairwise import cosine_similarity
    similarity_scores = cosine_similarity(pt)
    similarity_scores.shape
    
    sorted(list(enumerate(similarity_scores[70])),key = lambda x:x[1],reverse = True)
    
    print("-------------------------------------------------------------------------------------------------------------------------------")
    print("Pivot Table: ")
    print(pt)
    
    print("-------------------------------------------------------------------------------------------------------------------------------")
    print("final_ratings: ")
    print(final_ratings)
    
    print("-------------------------------------------------------------------------------------------------------------------------------")
    print("similarity Scores: ")
    print(similarity_scores)
    
    return pt,final_ratings,similarity_scores
```

### Arts

```python
df_Arts = pd.read_csv(r'D:\Academics\sem 4\Labs\ML_Mini-Project\new_phase\preprocessed_Arts.csv')
```

```python
arts_pt,arts_final_ratings,arts_similarity_scores = recommendations(df_Arts)
```
